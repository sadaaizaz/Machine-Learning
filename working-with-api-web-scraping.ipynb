# What is an API?

An API, or Application Programming Interface, is a set of rules that allows different software applications to communicate and share data with one another. They are data pipelines that transfer data from one point to another.

//TMDB API

```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=2309c13c9b1c2a8e1f325c7496e5bb2c')
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
```

```python
temp_df.head()
```

```python
df = pd.DataFrame()
```

```python
frames = []
for i in range(1, 429):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=2309c13c9b1c2a8e1f325c7496e5bb2c&page={i}')
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    frames.append(temp_df)
df = pd.concat(frames, ignore_index=True)
```

```python
df
```

```python
df.shape
```

```python
df.to_csv('movies.csv')
```

# Web Scraping

When API is not available, the web is scraped for pages. Web scraping is the automated process of extracting data from websites using bots or scripts. It fetches a page’s HTML code, isolates specific elements (like prices or text), and exports the information into structured formats like CSV or databases.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
pip install beautifulsoup4 lxml
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
    "Referer": "https://google.com",
    "Connection": "keep-alive"}
webpage = requests.get('https://books.toscrape.com/', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
for i in soup.find_all('h3'):
    print(i.text)
```

```python
company = soup.find_all('h3')
len(company)
```

```python
name=[]
for i in company:
    name.append((i).text)
```

```python
name
```

```python
d = {'Name':name}
df = pd.DataFrame(d)
df
```
