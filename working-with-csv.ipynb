# 3. Opening a csv file from a URL (copy paste)

# 12. Converters

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('/kaggle/input/datasets/sadaaizaz/ipldates/IPL Matches 2008-2020.csv', converters={'team1':rename})
```
